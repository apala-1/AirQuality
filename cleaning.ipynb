# Data Cleaning

This is the cleaning process of the Air Quality Dataset in Kathmandu Valley from Jan 2023 to March 2026.

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/kathmandu-air-quality.csv')
print(data.head())
```

```python
print(data.columns)
data.columns = (
    data.columns.astype(str)
                .str.strip()
                .str.lower()
)
print(data.columns)
```

```python
# First, i'll check each column

# print(data["date"].value_counts())
# print(data["pm25"].value_counts())
# print(data["pm10"].value_counts())

data['pm25'] = pd.to_numeric(data['pm25'], errors='coerce')
data['pm10'] = pd.to_numeric(data['pm10'], errors='coerce')

print(data.isnull().sum())
```

```python
data = data.dropna(subset=["pm25"])

print(data.isnull().sum())

data['date'] = pd.to_datetime(data['date'])
print(data['date'].head())

data = data[['date', 'pm25']]
```

```python
data.to_csv('data/cleaned_aqi.csv', index=False)
```
